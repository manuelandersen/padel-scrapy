```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.padelfip.com/es/ranking-masculino/'
response = requests.get(url)

if response.status_code == 200:
    html_content = response.text

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the element with class .section__sliderHome
    section_slider_home = soup.select_one('.section__sliderHome')

    if section_slider_home:
        # Find the element with class .slider__rankings.slick-initialized.slick-slider.slick-dotted inside .section__sliderHome
        slider_rankings = section_slider_home.select_one('.slider__rankings')

        if slider_rankings:
            # Print or process the HTML content of the matched element
            print(slider_rankings.prettify())  # This prints the element with proper indentation
        else:
            print("No element found with class .slider__rankings.slick-initialized.slick-slider.slick-dotted inside .section__sliderHome")
    else:
        print("No element found with class .section__sliderHome")
else:
    print(f"Failed to retrieve content. Status code: {response.status_code}")
```

```python
import requests
from lxml import html

url = 'https://www.padelfip.com/es/ranking-masculino/'
response = requests.get(url)

if response.status_code == 200:
    # Parse HTML content
    tree = html.fromstring(response.content)

    # Find element with class 'slider__number'
    slider_number_element = tree.xpath('//span[@class="slider__number"]')

    if slider_number_element:
        # Extract text content from the found element
        slider_number_text = slider_number_element[0].text_content().strip()
        print(slider_number_text)
    else:
        print("No element found with class .slider__number")
else:
    print(f"Failed to retrieve content. Status code: {response.status_code}")
```
